```python
# Identify outliers in BMI using IQR method
Q1 = df['BMI'].quantile(0.25)
Q3 = df['BMI'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

bmi_outliers = df[(df['BMI'] < lower) | (df['BMI'] > upper)]
bmi_outliers.head()
```

```python
# Box plot to visualize BMI outliers
plt.boxplot(df['BMI'])
plt.title("BMI Outliers", fontsize=16, fontweight='bold',)
plt.grid(axis='y', alpha=0.3)
plt.show()

```

```python
# Cleaning BMI outliers
df_clean = df[(df['BMI'] >= lower) & (df['BMI'] <= upper)]
# Compare number of rows before and after removing outliers
print("Original rows:", len(df))
print("After removing outliers:", len(df_clean))
# outliers were retained for risk modeling
```

```python
# Building a risk score combining multiple factors
risk_df = df.copy()

# Now create individual risk components
risk_df['BMI_risk'] = np.where(risk_df['BMI'] >= 30, 1, 0)
risk_df['Glucose_risk'] = np.where(risk_df['BloodGlucose'] >= 100, 1, 0)
risk_df['TG_risk'] = np.where(risk_df['Triglycerides'] >= 150, 1, 0)
risk_df['HDL_risk'] = np.where((risk_df['Sex'] == 'Male') & (risk_df['HDL'] < 40) |
                              (risk_df['Sex'] == 'Female') & (risk_df['HDL'] < 50), 1, 0)
risk_df['Waist_risk'] = np.where(
    ((risk_df['Sex'] == 'Male') & (risk_df['WaistCirc'] >= 102)) |
    ((risk_df['Sex'] == 'Female') & (risk_df['WaistCirc'] >= 88)), 1, 0)

risk_factors = [
    'BMI_risk',
    'Waist_risk',
    'Glucose_risk',
    'TG_risk',
    'HDL_risk'
]

risk_df['RiskScore'] = risk_df[risk_factors].sum(axis=1)
risk_df[['BMI_risk', 'Waist_risk', 'Glucose_risk', 'TG_risk', 'HDL_risk', 'RiskScore']].head()

# Categorizing risk levels according to risk scores
conditions = [
    (risk_df['RiskScore'] == 0),
    (risk_df['RiskScore'] >= 1) & (risk_df['RiskScore'] <= 2),
    (risk_df['RiskScore'] >= 3) & (risk_df['RiskScore'] <= 4),
    (risk_df['RiskScore'] >= 5)
]
choices = ['No Risk','Low Risk', 'Moderate Risk', 'High Risk']
risk_df['RiskLevel'] = np.select(conditions, choices, default='Unknown')
risk_df[['RiskScore', 'RiskLevel']].head()
```

```python
# Save the final dataset with risk scores
risk_df.to_csv('D:\Python_SQL Project\my_env1\\Metabolic_Syndrome_With_RiskScores.csv', index=False)
```
